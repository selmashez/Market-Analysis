to import essential libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import colorama
from colorama import Fore , Style
```

to read data and create the DataFrame 

```python
file = pd.read_csv('Market.csv')
file
```

```python
df = pd.DataFrame(file)
df
```

choosing NYA Indexes

```python
df = df[df.Index == 'NYA']
df
```

With this code, all numbers in the columns will become integers.

```python
def make_int(x):
    if (x - np.floor(x)) >= 0.5:
        return np.ceil(x)
    else:
        return np.floor(x)
    

df['Open'] = df['Open'].apply(make_int)
df['High'] = df['High'].apply(make_int)
df['Low'] = df['Low'].apply(make_int)
df['Close'] = df['Close'].apply(make_int)
df['Adj Close'] = df['Adj Close'].apply(make_int)

df
```

Removing Noise Datas

```python
df = df.drop(index=[829,831,833])
df
```

Let's have the number of  missing values of each column

```python
for column in df:
    null_count = df[column].isnull().sum()
    print(f'The number of missing values in {column} column is: {Fore.RED}{null_count}{Style.RESET_ALL}')
```

Since the number of missing values in the data is very low, in my opinion, removing them is the most sensible approach.

```python
# Removing missing values:

df = df.dropna()
df
```

Let's have Linear analysis for date and Adj Close Using diffrent plots

```python
plt.figure(figsize=(10, 5)) 
plt.plot(df['Date'], df['Adj Close'], marker='o', linestyle='-', color='r')
plt.show()
```

```python
df['Date_ordinal'] = df['Date'].apply(lambda date: date.toordinal())
plt.figure(figsize=(12, 6))
sns.residplot(x = 'Date_ordinal', y = 'Adj Close', data = df , lowess=True, color="g")
```
